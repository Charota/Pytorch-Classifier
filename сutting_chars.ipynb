```python
path_aida_classification = '/home/user/Desktop/NNPhotomath/dataSets/aida_classification'
```

```python
for i in range(len(jsons_links)):
    json_file = open(jsons_links[i])   #i is batch number
    json_entity = json.load(json_file) #len(json_entity) => 10 000
    
    batch_len = 1000
    last_batch_num = 0
    new_element_num = 0
    old_batch_num = 0
    
    for j in range(len(json_entity)):
        img = cv2.imread(f'/home/user/Desktop/NNPhotomath/dataSets/aida_calc_dataset/batch_{i+1}/background_images/{json_entity[j]["filename"]}')
        
        if img is not None:
            for k in range(len(json_entity[j]["image_data"]["visible_latex_chars"])):
                char_name = json_entity[j]["image_data"]["visible_latex_chars"][k]
                char_xmin = json_entity[j]["image_data"]["xmins_raw"][k]
                char_xmax = json_entity[j]["image_data"]["xmaxs_raw"][k]
                char_ymin = json_entity[j]["image_data"]["ymins_raw"][k]
                char_ymax = json_entity[j]["image_data"]["ymaxs_raw"][k]

                cropped_char = img[char_ymin:char_ymax, char_xmin:char_xmax]
            
                cropped_char = cv2.resize(cropped_char, (32, 32), interpolation= cv2.INTER_LINEAR)

                new_element_num += 1
                if new_element_num > batch_len:
                    new_element_num = 0
                    last_batch_num += 1
                
                path_exists_aida = os.path.exists(path_aida_classification)
                if not path_exists_aida:
                    os.mkdir(path_aida_classification)
                
                path_exists_batch_num = os.path.exists(f'{path_aida_classification}/{last_batch_num}')
                if old_batch_num != last_batch_num:
                    if not path_exists_batch_num:
                        os.mkdir(f'{path_aida_classification}/{last_batch_num}')
                
                
                cv2.imwrite(f'/home/user/Desktop/NNPhotomath/dataSets/aida_classification/{last_batch_num}/{json_entity[j]["image_data"]["visible_latex_chars"][k]}_{chars_amount[json_entity[j]["image_data"]["visible_latex_chars"][k]]}.jpg', cropped_char)
                chars_amount[json_entity[j]["image_data"]["visible_latex_chars"][k]] += 1
                
                
```
